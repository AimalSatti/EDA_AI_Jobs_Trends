 # *1.  Which AI job roles command the highest salary premiums compared to their base category? Are LLM roles consistently paying more than other AI roles?*



```python
# Question 1: Salary Premium by Job Title
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top 10 jobs by average salary premium
salary_premium_by_role = df.groupby('job_title')['ai_salary_premium_pct'].agg(['mean', 'count']).sort_values('mean', ascending=False).head(10)
salary_premium_by_role.plot(kind='barh', y='mean', ax=axes[0], color='steelblue', legend=False)
axes[0].set_title('Top 10 Job Roles by Average Salary Premium (%)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Average Salary Premium (%)')
axes[0].invert_yaxis()

# Salary premium distribution by job category
category_premium = df.groupby('job_category')['ai_salary_premium_pct'].mean().sort_values(ascending=False)
category_premium.plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Average Salary Premium by Job Category', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Average Salary Premium (%)')
axes[1].set_xlabel('Job Category')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('01_salary_premium_by_role.png', dpi=300, bbox_inches='tight')
plt.show()

print("Question 1 Results:")
print("\nTop 10 Jobs by Salary Premium:")
print(salary_premium_by_role[['mean']])
```

# *2. How does salary vary by experience level and job category?*

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

#  Salary by experience level
exp_salary = df.groupby('experience_level')['annual_salary_usd'].agg(['mean', 'median', 'count']).sort_values('mean')
exp_salary[['mean', 'median']].plot(kind='bar', ax=axes[0], color=['#3498db', '#e74c3c'])
axes[0].set_title('Average & Median Salary by Experience Level', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Salary (USD)')
axes[0].set_xlabel('Experience Level')
axes[0].legend(['Mean', 'Median'])
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
axes[0].tick_params(axis='x', rotation=45)

#  Heatmap: Experience vs Category
pivot_salary = df.pivot_table(values='annual_salary_usd', 
                               index='experience_level', 
                               columns='job_category', 
                               aggfunc='mean')
sns.heatmap(pivot_salary, annot=True, fmt='.0f', cmap='YlGnBu', ax=axes[1], cbar_kws={'label': 'Avg Salary (USD)'})
axes[1].set_title('Average Salary: Experience Level vs Job Category', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('03_salary_by_experience_category.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nQuestion 2 Results:")
print("\nAverage Salary by Experience Level:")
print(exp_salary[['mean', 'median']])
```
